```python
fig = plt.figure(figsize=(18, 30)) # any figure has 10x10 size
fig_dims = (5, 2) # 4 rows and 5 collums 
b = 17
plt.subplot2grid(fig_dims, (0, 0))
df['A'].plot(kind='kde')
plt.title('A')

plt.subplot2grid(fig_dims, (0, 1))
df['B'].plot(kind='kde')
plt.title('B')

plt.subplot2grid(fig_dims, (1, 0))
df['C'].plot(kind='kde')
plt.title('C')

plt.subplot2grid(fig_dims, (1, 1))
df['D'].plot(kind='kde')
plt.title('D')

plt.subplot2grid(fig_dims, (2, 0))
df['E'].plot(kind='kde')
plt.title('E')

plt.subplot2grid(fig_dims, (2, 1))
df['F'].plot(kind='kde')
plt.title('F')

plt.subplot2grid(fig_dims, (3, 0))
df['G'].plot(kind='kde')
plt.title('G')

plt.subplot2grid(fig_dims, (3, 1))
df['H'].plot(kind='kde')
plt.title('H')

plt.subplot2grid(fig_dims, (4, 0))
df['I'].plot(kind='kde')
plt.title('I')

plt.subplot2grid(fig_dims, (4, 1))
df['L'].plot(kind='kde')
plt.title('L')

plt.show()
plt.savefig('../Images/feature_distribution.png')
```

```python
f = plt.figure(figsize=(10, 10))
plt.matshow(df.corr(), fignum=f.number)
plt.xticks(range(df.shape[1]), df.columns, fontsize=14, rotation=45)
plt.yticks(range(df.shape[1]), df.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16)
plt.savefig('../Images/Correlation.png')
```
